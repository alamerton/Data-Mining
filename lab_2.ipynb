```python
import data
import matplotlib as plt
import pandas as pd
import numpy as np
```

Convert CSV to dataframe

```python
df = pd.read_csv("data/london-borough-profiles-jan2018.csv", encoding='unicode_escape')
df.head()
```

```python
print(df.columns)
```

Create new dataframe containing the target values

```python
new_df = df[['Male life expectancy, (2012-14)','Female life expectancy, (2012-14)']].copy()
```

Check contents of columns of new dataset

```python
column1 = new_df['Male life expectancy, (2012-14)']
column2 = new_df['Female life expectancy, (2012-14)']

print(column1)
print(column2)
```

Remove NaN values from dataset

```python
new_df.dropna(0)
# new_df['Male life expectancy, (2012-14)'].replace('.', '', inplace=True)
# new_df['Female life expectancy, (2012-14)'].replace('.', '', inplace=True)
new_df['Male life expectancy, (2012-14)'].replace(np.nan, '', inplace=True)
new_df['Female life expectancy, (2012-14)'].replace(np.nan, '', inplace=True)

new_df['Female life expectancy, (2012-14)'].drop('.')
new_df['Male life expectancy, (2012-14)']
new_df['Male life expectancy, (2012-14)'].drop('.')
```

```python
column1 = new_df['Male life expectancy, (2012-14)']
column2 = new_df['Female life expectancy, (2012-14)']

print(column1)
print(column2)
```

```python
new_df['Male life expectancy, (2012-14)'] = new_df['Male life expectancy, (2012-14)'].astype(int)
new_df['Female life expectancy, (2012-14)'] = new_df['Female life expectancy, (2012-14)'].astype(int)
```
